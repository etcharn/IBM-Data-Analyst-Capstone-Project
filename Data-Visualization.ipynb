**Connect to the database**

```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

**Import pandas module**

```python
import pandas as pd
```

**Basics of running sql query**

```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

# Set options to display all columns and rows (if needed)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

df.head(5)
```

**Listing all tables in the database**

```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

**Boxplot of Age**

```python
import seaborn as sns
import matplotlib.pyplot as plt

QUERY = """
SELECT Age
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

# Create the boxplot
sns.boxplot( 
    y = "Age",  # Column with numerical values
    data = df
)

# Display the plot
plt.show()
```

### Visualizing relationships in data

**Scatter plot of Age and WorkWeekHrs**

```python
# query Age and WorkWeekHrs from master table
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

# Have a brief look at their correlations
df['Age'].corr(df['WorkWeekHrs'])
```

**The correlation doesn't give much of an idea how the two values related**

```python
# Create a scatter plot
plt.scatter(df['Age'], df['WorkWeekHrs'])

# Add labels and title
plt.xlabel("X-Values")
plt.ylabel("Y-Values")
plt.title("Scatter Plot")

# Show the plot
plt.show()
```

**Even with the scatterplot, it's hard to conclude their relation**

### Visualizing composition of data

**Create a pie chart of the top 5 databases that respondents wish to learn next year.**

```python
# Beginning with query data from the associated database
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

# see the top 5 role of query data
df.head(5)
```

**Count the number of occurrence of each database**

```python
df['DatabaseDesireNextYear'].value_counts()
```

**Extract the top 5 rows**

```python
# Extract top 5 rows of database count
df_piechart = df['DatabaseDesireNextYear'].value_counts().head(5)

# create a dataframe out of it
df_piechart = pd.DataFrame(df_piechart)
# assign column name
df_piechart.columns = ['Count']

df_piechart
```

**Plot the pie chart**

```python
# Create a pie chart
plt.figure(figsize=(10, 10))  # Adjust figure size as needed

# Create a list of label out of the index of the dataframe
labels = df_piechart.index.to_list()

# Plot the pie chart with Count 
plt.pie(df_piechart['Count'], labels=labels, autopct="%1.1f%%")  
plt.legend(labels, title="Database") 

# Add title
plt.title("Pie Chart of Sample Data")

# Show the plot
plt.show()
```

**Find the number of respondents who indicated that they work on 'MySQL' only?**

```python
# Query the associated table
QUERY = """
SELECT *
FROM DatabaseWorkedWith
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)

# View the top 5 roles of the dataframe
df.head(5)
```

**If a respondent only work on SQL, there wouldn't be a repetitive Repondent_id in the dataframe as the Respondent_id only exist in a row**

```python
# Remove all duplicate rows with repetitive Respondent_id
df_noDupRes = df.drop_duplicates(subset='Respondent', keep=False)

# Find all the rows having SQL as DatabaseWorkedWith
df_noDupRes.loc[df_noDupRes.DatabaseWorkedWith == 'MySQL'].shape[0]
```

**Plot the median ConvertedComp for all ages from 45 to 60**
